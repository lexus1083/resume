```python
import matplotlib.pyplot as plt  #загружаем библиотеку matplotlib
import pandas as pd   #загружаем библиотеку pandas

```

```python
from google.colab import files
upload = files.upload()  #создаем возможность загрузки файла в notebook
```

```python
import io
wc = pd.read_csv(io.BytesIO(upload['FIFA_World_Cup_1558_23.csv']))  #через импорт io визуализируем таблицу в notebook
wc
```

```python
counter = 0
for i in wc['Home.Team.Initials']:  #возвращаем количество совпадений конкретного элемента списка
    if i == 'GER':
        counter +=1
print(counter)
```

```python
wc['Home.Team.Initials'].value_counts()  #возвращаем количество уникальных совпадений в колонке Home.Team.Initials (Бразилия чаще всех играла дома)
```

```python
plt.title('Посещаемость')   #называем таблицу
plt.xlabel('Year')   #называем ось х
plt.ylabel('Attendance')   #называем ось у
plt.bar(wc['Year'], wc['Attendance'])  #загружаем из таблицы данны в ось х и ось у
plt.show ()   #показываем результат
```

```python
plt.title('Голы дома')   #называем таблицу
plt.xlabel('Year')   #называем ось х
plt.ylabel('Home.Team.Goals')   #называем ось у
plt.bar(wc['Year'], wc['Home.Team.Goals'])  #загружаем из таблицы данны в ось х и ось у
plt.show ()   #показываем результат
```
